```python
from IPython.display import Image, display
import os
import glob

WORKLOAD = 'resnet50_selected'
BASE_PATH = '/home/workspace'

relative_plots_dir_in_workspace = f"evaluation_setups/structured_sparsity/outputs/{WORKLOAD}/plots"
plots_dir = os.path.join(BASE_PATH, relative_plots_dir_in_workspace)

plot_files = glob.glob(os.path.join(plots_dir, '*.png'))

if plot_files:
    plot_files.sort()
    for image_filename in plot_files:
        print(f"\nDisplaying: {os.path.basename(image_filename)}")
        display(Image(filename=image_filename))
else:
    print(f"ERROR: No .png files found in the directory.")
    print("Please ensure the comparison script ran successfully and generated plots.")
```
